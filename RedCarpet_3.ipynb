```python
loan_data = xls["Loan data"]
loan_data['due_date'] = pd.to_datetime(loan_data['due_date'] )

loan_data
```

```python
from datetime import datetime, timedelta,date
loan_data_3m = loan_data[(loan_data.due_date >= date(2018,10,1)) & (loan_data.due_date <= date(2019,1,1))]
loan_data_3m
```

```python
loan_data_3m['payment_date'] = pd.to_datetime(loan_data_3m['payment_date'] )
loan_data_3m["payment_date"] = [d.date() for d in loan_data_3m["payment_date"]]
loan_data_3m['payment_date'] = pd.to_datetime(loan_data_3m['payment_date'] )
# loan_data_3m['payment_date'] = pd.to_datetime(loan_data_3m['payment_date'] , format = '%d/%m/%y %H:%M' )

loan_data_3m
```

```python
a=[] 
b=[]
for j in loan_data_3m['payment_date']:
    a.append(j.month)
    b.append(13-j.month)
loan_data_3m['payment_month'] = a
loan_data_3m['rec_ref'] = b
loan_data_3m['payment_amount_copy'] = loan_data_3m['payment_amount']
loan_data_3m
# def monthdel(m):
#     for i in m
#     delta = 13 - i
#     if (delta > 3):
#         delta = 3
#     return delta
```

```python
from dateutil import relativedelta 
from calendar import monthrange
from datetime import datetime, timedelta
# date1 = date(2019,1,1)

# 'payment_date': lambda datetime: (date1 - datetime.max()),
rfm= loan_data_3m.groupby('cust_id').agg({'rec_ref': lambda dt: min(dt),
                                        'payment_amount_copy': lambda num: len(num),
                                        'payment_amount': lambda price: price.sum()})
rfm.columns
rfm.columns=['recency','frequency','Monetary']
rfm['Recency']=""
rfm['Frequency'] = ""
rfm.recency = rfm.recency.replace(np.nan , 4)
j = []
k=[]
for i in rfm['recency']:
    if ((i > 3) or (i == np.nan)):
        j.append(4)
    if( i <=3):
        j.append(i)
rfm['Recency']=j
for i in rfm['frequency']:
    if ((i > 3) or (i == np.nan)):
        k.append(3)
    if( i <=3):
        k.append(i)
rfm['Frequency'] = k
rfm = rfm.drop('recency',axis = 1)
rfm = rfm.drop('frequency',axis = 1)
rfm
```

```python
# rfm['r_quartile'] = pd.qcut(rfm['Recency'], 4)
# rfm['f_quartile'] = pd.qcut(rfm['Frequency'], 3)
# rfm['m_quartile'] = pd.qcut(rfm['Monetary'], 4)
```

```python
rfm['r_rank'] = rfm['Recency'].rank(method='first')
rfm['f_rank'] = rfm['Frequency'].rank(method='first')

rfm['r_quartile'] = pd.qcut(rfm['r_rank'], 4, ['1','2','3','4'])
rfm['f_quartile'] = pd.qcut(rfm['f_rank'], 4, ['4','3','2','1'])
rfm['m_quartile'] = pd.qcut(rfm['Monetary'], 4, ['4','3','2','1'])

rfm = rfm.drop('r_rank',axis = 1)
rfm = rfm.drop('f_rank',axis = 1)
rfm['RFM_Category'] = rfm.r_quartile.astype(str)+ rfm.f_quartile.astype(str) + rfm.m_quartile.astype(str)
#rfm['Score']=""
rfm['Score'] = int(rfm.r_quartile.astype(str)) + int(rfm.f_quartile.astype(str)) + int(rfm.m_quartile.astype(str))
rfm.dtypes
```
